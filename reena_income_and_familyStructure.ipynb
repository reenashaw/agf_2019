```python
unders_income_rows_df= fsa_df.loc[unders_income_rows, ] #loc,and not iloc
# includes: 'under $ 10,000' and 'under 5 years'
# to filter only by $
unders_income_rows_df
```

```python
fsa_char= fsa_df['characteristic']
dollar_income_rows= fsa_char.str.startswith('$' )  # stores boolean values for each record.
# syntax: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html

#unders_income_rows  # works
dollar_income_rows_df =fsa_df.loc[dollar_income_rows, ] # 114410 *7
dollar_income_rows_df
```

```python
# join the 2 income df's
income_dfs_list= [unders_income_rows_df, dollar_income_rows_df]
income_df= pd.concat(income_dfs_list)
#income_df.head()
income_df.shape # 122200*7
```
